```python
# leer csv
df = pd.read_table("data/air_temp.2014", sep="\s+", header=None)
```

```python
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sept","Oct","Nov","Dic"]
```

```python
def convertir_fa_to_c(f):
    return (f-32) * 0.5556
```

```python
for mes in meses:
    df["{}_c".format(mes)] = df.apply(lambda x: convertir_fa_to_c(x[mes]), axis=1)
```
